```python
from MLOps_project.utils.common import read_yaml,createDIr
```

```python
import urllib.request as request
from MLOps_project import logger
import zipfile
```

```python
class Config_manager:
    def __init__(self):
      
        config_path = Path('config/config.yaml')
        params_path = Path('params.yaml')
        schema_path = Path('schema.yaml')

        
        self.config = read_yaml(config_path)
        self.params = read_yaml(params_path)
        self.schema = read_yaml(schema_path)

        createDIr([self.config.artifact_folder_loc])

    
    def create_root_down(self):
        config = self.config.data_ingestion
        createDIr([config.rootdir_zip_down])

    def download_the_file(self):
        config = self.config.data_ingestion
        if not os.path.exists(config.zip_downl_loc):
            filename,header=request.urlretrieve(url=config.source_url , filename=config.zip_downl_loc )

            logger.info(f'the file {filename} has successfully downloaded')
        else:
            logger.info(f'file already exists')

    def unzip_data(self):
        config = self.config.data_ingestion
        unzip_path= config.unzip_dir_loc
        os.makedirs(unzip_path,exist_ok=True)
        with zipfile.ZipFile(config.zip_downl_loc,'r') as zip_ref:
            zip_ref.extractall(unzip_path)
        logger.info(f'data has successfully extracted')
```

```python
try:
    config = Config_manager()
    config.create_root_down()
    config.download_the_file()
    config.unzip_data()
except Exception as e:
    raise e
```
